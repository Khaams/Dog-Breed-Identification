```python
import os

IMAGES_DIR = './data/Images/'
breed_list = os.listdir(IMAGES_DIR)

print(f"Il y a {len(breed_list)} races")

images_by_breed = {breed.split('-')[-1] : len(os.listdir(f"{IMAGES_DIR}/{breed}")) for breed in breed_list}
total_images = sum([v for k,v in images_by_breed.items()])

print(f"Il y a un total de {total_images} d'images")
```

```python
#### import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(font_scale = 2)

names = list(images_by_breed.keys())
values = list(images_by_breed.values())

plt.figure(num=None, figsize=(20,18), dpi=80, facecolor='w', edgecolor='r')
ax = sns.barplot(x = names, y = values,orient='v')

ax.set_ylabel('Images par races', fontsize = 11)
ax.set_xlabel('Nom de race', fontsize = 11)

plt.xticks(rotation = 90, fontsize=11)
plt.yticks(fontsize=11)

plt.show()
```

```python
def show_dir_images(breed, n_to_show):
    plt.figure(figsize=(16,16))
    img_dir = f"{IMAGES_DIR}/{breed}/"
    images = os.listdir(img_dir)[:n_to_show]
    for i in range(n_to_show):
        img = plt.imread(img_dir + images[i])
        plt.subplot(int(n_to_show/4+1), 4, i+1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f'{breed}', fontsize = 10)
```

```python
print(breed_list[42])
show_dir_images(breed_list[42], 4)
```
